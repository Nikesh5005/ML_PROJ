```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
X_train=pd.read_csv("C:\\Users\\patha\\OneDrive\\Desktop\\ML\\ML\\ML_PROJ\\X_train")
X_test=pd.read_csv("C:\\Users\\patha\\OneDrive\\Desktop\\ML\\ML\\ML_PROJ\\X_test")
y_train=pd.read_csv("C:\\Users\\patha\\OneDrive\\Desktop\\ML\\ML\\ML_PROJ\\y_train")
y_test=pd.read_csv("C:\\Users\\patha\\OneDrive\\Desktop\\ML\\ML\\ML_PROJ\\y_test")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
train_pradiction=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,train_pradiction)
train_acc
```

```python
pradiction=model.predict(X_test)
pradiction
```

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,pradiction)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pradiction)
cm
```

```python
def pradict_plot(target,x):
    prad=model.predict(x)
    f_accuracy=accuracy_score(target,prad)
    print(f_accuracy)
    confusion_mat=confusion_matrix(target,prad,normalize='true')
    sns.heatmap(confusion_mat,annot=True)
    #plt.xlabel('Prediction')
    #plt.ylabel('Target')
    plt.show()
```

```python
pradict_plot(y_test,X_test)
```
