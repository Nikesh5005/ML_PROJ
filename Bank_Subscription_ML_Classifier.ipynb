```python
plt.figure(figsize=(12,4))
sns.countplot(data=df,x='loan',hue='deposit')
plt.show()
```

From the analysis, it is evident that the majority of customers in the dataset do not have a loan, while only a small proportion of customers have an existing loan.
Among customers without a loan, the responses are fairly balanced, with a similar number of customers agreeing and disagreeing to subscribe to the term deposit. In contrast, for customers who already have a loan, a significantly higher proportion tend to decline the subscription, with only a small number showing interest.
This suggests that having an existing loan may negatively influence a customer’s likelihood of subscribing to the bank’s term deposit.


```python
plt.figure(figsize=(12,4))
sns.countplot(data=df,x='month',hue='deposit')
plt.show()
```

The analysis shows that customers are more likely to subscribe at the beginning (February–April) and end of the year (September–December). In contrast, during the middle months, customers are more likely to decline. Overall, “yes” responses are higher at the start and end of the year, while “no” responses dominate in the middle, indicating a seasonal trend in customer behavior.
